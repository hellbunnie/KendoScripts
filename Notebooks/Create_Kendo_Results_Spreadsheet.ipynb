## Step 1: specify the files
This step will ask you for the names of the uploaded registration files (leave blank for 2nd / 3rd file if they don't exist), and the results output file.

```python
import pandas as pd
from os import path

# Get the filename of the exam registration files
inputfile1 = input('Please enter the name of the first file you have uploaded containing the registrations: ')
inputfile2 = input('Please enter the name of the second file you have uploaded containing the registrations: ')
inputfile3 = input('Please enter the name of the third file you have uploaded containing the registrations: ')

# Get the filename of the exam results template file
outputfile = input('Please enter the name of the uploaded results template file: ')




```

##Step 2: Load and prepare the data
This step will read in the registration files and concatenate them together, then remove unneccesary fields

```python
# Read in the registration files

li = []
for filename in [inputfile1,inputfile2,inputfile3]:
    if path.isfile(filename):
      df = pd.read_excel(filename, index_col=None, header=0)
      li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
df.rename(columns={'Sex': 'Gender'}, inplace=True)

namesframe = df.drop(['Unnamed: 0', 'Reg. for Dan', 'EKFCode',
       'Birthday', 'Age', 'Member', 'reg. Dan', 'Created at', 'Created by',
       'Changed at', 'Changed by', 'Verified at', 'Verified by', 'Deleted at',
       'Deleted by', 'Email', 'ExamDate', 'Discipline', 'Country', 'Location',
       'Head of Jury', 'Organized by', 'Organized By Short', 'GenterTitle',
       'DanName', 'DanCode3', 'DanCode3.1', 'DanCode4', 'App.Fee', 'Reg.Fee'], axis=1)

```

## Step 3: Write the results file
This step will write out the data from the input files to the output files (names, gender to the Names sheet, and the exam #s to each of the grade sheets). It will apply the necessary conditional formatting to the Names sheet.

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from openpyxl.formatting.rule import FormulaRule

# 1. Write data to Excel
with pd.ExcelWriter(outputfile, engine="openpyxl", mode="a", if_sheet_exists="overlay") as writer:
    namesframe.to_excel(writer, sheet_name="Names", index=False, startrow=0, startcol=0)

# 2. Reopen workbook
wb = load_workbook(outputfile)
sheet = wb["Names"]

# 3. Define conditional formatting (green if D is not blank)
green_fill = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")
rule = FormulaRule(formula=['$E2<>""'], fill=green_fill)

# 4. Apply formatting to A & B for all data rows
max_row = len(df) + 1  # +1 for header row
sheet.conditional_formatting.add(f"A2:B{max_row}", rule)

# 5. Save
wb.save(outputfile)
```
