```python
!pip install reportlab
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4

def numbers_to_pdf_stickers(numbers, filename="numbers.pdf"):
    c = canvas.Canvas(filename, pagesize=A4)
    page_width, page_height = A4

    # Sticker layout: 2x2 per page
    sticker_width = page_width / 2
    sticker_height = page_height / 2

    # X,Y positions: center of each sticker
    positions = [
        (sticker_width/2, 2.9*sticker_height/2),  # Top-left
        (3*sticker_width/2, 2.9*sticker_height/2),  # Top-right
        (sticker_width/2, (sticker_height/2)-20),  # Bottom-left
        (3*sticker_width/2, (sticker_height/2)-20),  # Bottom-right
    ]

    for i, number in enumerate(numbers):
        pos_index = i % 4
        x, y = positions[pos_index]

        number_str = str(number)
        if len(number_str) != 3:
            raise ValueError("All numbers must be 3 digits")

        # First digit (top line) ~150pt
        c.setFont("Times-Roman", 160)
        c.drawCentredString(x, y + 60, number_str[0])

        # Last two digits (bottom line) ~200pt
        c.setFont("Times-Roman", 220)
        c.drawCentredString(x, y - 100, number_str[1:])

        # Start new page after 4 stickers
        if pos_index == 3:
            c.showPage()

    # Save the last page if it's not full
    if len(numbers) % 4 != 0:
        c.showPage()

    c.save()

# Example usage
numbers_to_pdf_stickers(numbers)
```
